```python
class dataframe():
    
    def __init__(self, file):
        self.start_year=str(file)[0:4]
        self.end_year=str(file)[5:9]
        self.df=pd.read_csv(f'C:/Users/Nate/Documents/OA3801 Computational Methods/Final Project/NHL_Data/CSV/{file}')
```

```python
list_df_objects=[]
for file in files:
    list_df_objects.append(dataframe(file))
list_df_objects[12].df
```

```python
for obj in list_df_objects:
    
    #input Team as a column name for files that are missing it
    obj.df.loc[0, 'Unnamed: 1']='Team'
    
    #make the first row the column names
    obj.df.columns = obj.df.iloc[0]
    obj.df=obj.df.drop(0, axis=0)
    
    #remove all unnecessary columns
    obj.df=obj.df[['Team','GP','W','L']]
    
    
    #drop the league avaerage row at the end of each dataframe
    obj.df=obj.df.set_index('Team')
    obj.df=obj.df.drop(index='League Average', axis=0)

    
    #add the start year and end year columns
    start_year_list=[]
    end_year_list=[]
    for i in range(len(obj.df)):
        start_year_list.append(obj.start_year)
        end_year_list.append(obj.end_year)
    obj.df['start_year']=start_year_list
    obj.df['end_year']=end_year_list
    obj.df['start_year']=obj.df['start_year'].astype('int')
    obj.df['end_year']=obj.df['end_year'].astype('int')
    
    #add the win percentage column
    obj.df['W']=obj.df['W'].astype('int')
    obj.df['GP']=obj.df['GP'].astype('int')
    
    obj.df['W-L%']=obj.df['W']/obj.df['GP']
    
    #reset the index
    obj.df=obj.df.reset_index()
    
    #remove the asterix from team name
    for i in range(len(obj.df['Team'])):
        obj.df['Team'][i]=obj.df['Team'][i].rstrip('*')

    
list_df_objects[12].df
```
