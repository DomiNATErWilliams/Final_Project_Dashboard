```python
# Final Project Code for MLB:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
MLB_Stats = pd.read_csv('MLB_Stats.csv')
MLB_Stats_0 = MLB_Stats.fillna(0)
MLB_Stats_0.head()
```

```python
MLB_WinLoss = MLB_Stats_0.iloc[:,2:].div(MLB_Stats['G'],axis=0)
MLB_WinLoss.insert(0, 'Year', MLB_Stats['Year'])
#MLB_WinLoss.set_index('Year',inplace=True)
MLB_WinLoss.head()
```

```python
Teams = MLB_WinLoss.columns.values.tolist()[1:]
#[MLB_WinLoss.columns[1:],MLB_WinLoss['Year']]
Teams
```

```python
MLB_melt = MLB_WinLoss.melt(id_vars=['Year'], value_vars = Teams, var_name='Tm', value_name='W-L%')
MLB_melt
```

```python
MLB_WL = MLB_melt.sort_values('Year', ascending = True)
MLB_df = MLB_WL[MLB_WL['W-L%'] != 0]
MLB_df = MLB_df[(MLB_df['Year'] != 1901) & (MLB_df['Year'] != 1902)]
MLB_df
```

```python
# Significant MLB Team Franchise moves:
MLB_df.loc[(MLB_df['Tm'] == 'SFG') & (MLB_df['Year'] <= 1957), 'Tm'] = 'NYG' # New York Giants to SF Giants

MLB_df.loc[(MLB_df['Tm'] == 'OAK') & (MLB_df['Year'] <= 1954), 'Tm'] = 'PHA' # Philadelphia Athletics 1903-1954
MLB_df.loc[(MLB_df['Tm'] == 'OAK') & (MLB_df['Year'] > 1954) & (MLB_df['Year'] <= 1967), 'Tm'] = 'KCA' # KC Athletics 1955-1967

MLB_df.loc[(MLB_df['Tm'] == 'ATL') & (MLB_df['Year'] <= 1952), 'Tm'] = 'BOB' # Boston Braves 1903-1952
MLB_df.loc[(MLB_df['Tm'] == 'ATL') & (MLB_df['Year'] > 1952) & (MLB_df['Year'] <= 1965), 'Tm'] = 'MIB' # Milwaukee Braves 1953-1965

MLB_df.loc[(MLB_df['Tm'] == 'MIN') & (MLB_df['Year'] <= 1960), 'Tm'] = 'WHS' # Washington Senators 1903-1960

MLB_df.loc[(MLB_df['Tm'] == 'LAD') & (MLB_df['Year'] <= 1957), 'Tm'] = 'BRK' # Brooklyn Dodgers 1903-1957

MLB_df.loc[(MLB_df['Tm'] == 'WSN') & (MLB_df['Year'] <= 2004), 'Tm'] = 'MOE' # Montreal Expos 1969-2004

```

```python
def convert_latitude(latitude_str):
    numerical_part = latitude_str.split('Â°')[0]  # Extract numerical part before 'Â°'
    numerical_value = float(numerical_part)
    
    if 'S' in latitude_str:
        numerical_value = -numerical_value  # If latitude is in Southern hemisphere
    
    return numerical_value

def convert_longitude(longitude_str):
    numerical_part = longitude_str.split('Â°')[0]  # Extract numerical part before 'Â°'
    numerical_value = float(numerical_part)
    
    if 'W' in longitude_str:
        numerical_value = -numerical_value  # If longitude is in Western hemisphere
    
    return numerical_value
```

```python
MLB_final['Latitude'] = MLB_final['Latitude'].apply(convert_latitude)
MLB_final['Longitude'] = MLB_final['Longitude'].apply(convert_longitude)
```

```python
import plotly.graph_objects as go
import pandas as pd
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

 

# Reset index after groupby
df = MLB_final.groupby(['Year', 'Tm']).sum().reset_index()

 

# Create a Dash web application
app = dash.Dash(__name__)

 

# Define the unique years for the dropdown options
year_options = [{'label': str(year), 'value': year} for year in df['Year'].unique()]

 

# Define the app layout
app.layout = html.Div([
    html.Div([
        dcc.Graph(id='animated-map'),
    ], style={'width': '70%', 'display': 'inline-block'}),
    html.Div([
        html.Label('Select Year:'),
        dcc.Dropdown(
            id='year-dropdown',
            options=year_options,
            value=df['Year'].min(),
        ),
        html.Div(id='slider-output-container'),
    ], style={'width': '20%', 'display': 'inline-block'}),
])

 

@app.callback(
    Output('animated-map', 'figure'),
    Output('slider-output-container', 'children'),
    Input('year-dropdown', 'value')
)
def update_map(selected_year):
    # Create an animated scatter plot using Plotly graph_objects
    fig = go.Figure()

 

    hover_text = []
    for index, row in df[df['Year'] == selected_year].iterrows():
        champion_text = "Champion" if row['Champion'] == 1 else "Not Champion"
        text = f"Team: {row['Tm']}<br>Year: {selected_year}<br>W-L%: {row['W-L%']}<br>{champion_text}"
        hover_text.append(text)

 

    scatter = go.Scattergeo(
        lon=df[df['Year'] == selected_year]['Longitude'],
        lat=df[df['Year'] == selected_year]['Latitude'],
        text=hover_text,
        mode='markers',
        marker=dict(
            size=df[df['Year'] == selected_year]['W-L%'] * 50,
            color=df[df['Year'] == selected_year]['W-L%'],
            colorscale='Viridis',
            cmin=0.3,
            cmax=0.7,
            colorbar=dict(title='Win-Loss Percentage'),
        ),
        name=str(selected_year),
    )
    fig.add_trace(scatter)

 

    fig.update_geos(
        scope='usa',
        showcoastlines=True,
        coastlinecolor="RebeccaPurple",
        showland=True,
        landcolor="LightGreen",
        showsubunits=True,
        subunitcolor="gray",
    )

 

    champion_team = df[(df['Year'] == selected_year) & (df['Champion'] == 1)]['Tm'].values
    if champion_team:
        champion_text = f'Champion: {champion_team[0]}'
    else:
        champion_text = 'No Champion'

 

    fig.update_layout(
        title=f'MLB Win-Loss Percentage Over Time - {selected_year}<br>{champion_text}',
        title_x=0.5
    )

 

    return fig, f'Selected Year: {selected_year}'

 

if __name__ == '__main__':
    app.run_server(debug=True, port = 8052)
```
